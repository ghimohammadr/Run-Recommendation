```python
import pandas as pd
import numpy as np
import folium
import matplotlib.pyplot as plt


hist_df = pd.read_excel('historical runs.xlsx')

hist_df = hist_df.dropna(subset='Round').reset_index(drop = True)
hist_df = hist_df[hist_df['Planned Distance'] != 'NaNkm'].reset_index(drop = True)

hist_df[['completed Drops', 'Total Drops']] = hist_df['Completed'].str.split('/', expand=True)

# Converting to numeric type if needed
hist_df[['completed Drops', 'Total Drops']] = hist_df[['completed Drops', 'Total Drops']].apply(pd.to_numeric)

hist_df = hist_df[hist_df['completed Drops'] != 0].reset_index()
hist_df = hist_df[~hist_df['completed Drops'].isna()].reset_index()


hist_df['distance'] = hist_df['Planned Distance'].str.replace('km', '', regex=True).astype(int)

hist_df['KM Cost'] = (hist_df['distance'] - 200) * 1
# Ensuring 'KM Cost' is not negative
hist_df['KM Cost'] = np.where(hist_df['KM Cost'] < 0, 0, hist_df['KM Cost'])

hist_df['Drop Cost'] = (hist_df['completed Drops'] - 70) * 4.5
# Ensuring 'KM Cost' is not negative
hist_df['Drop Cost'] = np.where(hist_df['Drop Cost'] < 0, 0, hist_df['Drop Cost'])

# Ensure the column is in datetime format
hist_df['Pln Start Date'] = pd.to_datetime(hist_df['Pln Start Date'], errors='coerce')

# Create a new column for the day of the week
hist_df['Day of Week'] = hist_df['Pln Start Date'].dt.day_name()

hist_df['Flat rate Cost'] = 370
hist_df['Total Cost'] = hist_df['KM Cost'] + hist_df['Drop Cost']+ hist_df['Flat rate Cost']
hist_df['Cost per Drop'] = hist_df['Total Cost']/hist_df['completed Drops']



# print('No. of runs:' , len(hist_df.groupby(['Round', 'Pln Finish Date']).count()))
# print('No. of drops:' , int(hist_df['completed Drops'].sum()))
# print('Max Km Bonus cost:' , hist_df['KM Cost'].max())
# print('Min Km Bonus cost:' , hist_df['KM Cost'].min())
# print('Avg Km Bonus cost:' , round(hist_df['KM Cost'].mean(), 2))
# print('Max Drop Bonus cost:' , hist_df['Drop Cost'].max())
# print('Min Drop Bonus cost:' , hist_df['Drop Cost'].min())
# print('Avg Drop Bonus cost:' , round(hist_df['Drop Cost'].mean(), 2))
```

```python
# Generate unique colors using 'gist_ncar' for more diversity
route_ids = sorted(drops_metro['RouteId'].unique())  # Sorting to maintain order
num_routes = len(route_ids)
colors = plt.cm.gist_ncar(np.linspace(0, 1, num_routes))  # More distinguishable colors

# Convert colors to hex values
route_colors = {route: f'#{int(c[0]*255):02x}{int(c[1]*255):02x}{int(c[2]*255):02x}' for route, c in zip(route_ids, colors)}

# Initialize the map
m = folium.Map(location=[drops_metro['Latitude'].mean(), drops_metro['Longitude'].mean()], zoom_start=12)

# Add points with unique colors
for _, row in drops_metro.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=5,
        color=route_colors[row['RouteId']],
        fill=True,
        fill_color=route_colors[row['RouteId']],
        fill_opacity=0.7,
        popup=f"Route ID: {row['RouteId']}"
    ).add_to(m)

# Add a legend manually with a scrollbar for large lists
legend_html = '''
<div style="position: fixed; 
            bottom: 50px; left: 50px; width: auto; height: auto; 
            overflow-y: auto; background-color: white; z-index:9999; 
            padding: 10px; font-size:14px; border-radius:5px; 
            box-shadow: 0px 0px 5px gray;">
    <b>Route ID</b><br>
'''
for route, color in route_colors.items():
    legend_html += f'<div style="display: flex; align-items: center;"><div style="width: 12px; height: 12px; background: {color}; margin-right: 5px;"></div>{route}</div>'

legend_html += '</div>'

m.get_root().html.add_child(folium.Element(legend_html))

# Display map
m
```
